# DOMAIN 1-Laut

```python
#file 1_65 LAUT
for time in range(0,73):
    xx = int(time)
    RH1_65 = getvar(file1_65,"rh",timeidx=xx)
    P1_65 = getvar(file1_65,"p",units='hpa',timeidx=xx)
    Z1_65 = getvar(file1_65,"z",units='m',timeidx=xx)
    
    Pl1_65 = P1_65[:,40,50]
    RHl1_65 = RH1_65[:,40,50]
    Zl1_65 = Z1_65[:,40,50]
    
    fig = plt.figure(figsize=(9,12))
#add_metpy_logo(fig, 115, 100)
    ax = plt.axes()

    ax.plot(RHl1_65, Zl1_65, 'g')
    #skew.plot(Pl1_11, thv1_11, 'g')
#skew.plot_barbs(p, u, v)
    ax.set_ylim(0,22000)
    ax.set_xlim(0, 100)
    ax.set_ylabel('Height (m)', fontsize=14)
    ax.set_xlabel('Relative Humidity (%)',fontsize=14)

    #skew.ax.legend(["Temperature", "Dew Point Temperature"],loc=0)
    plt.grid(True)
    plt.show()
    
    fig.savefig('plot/RH/laut/d01_rh99_laut_ke_'+str(xx) +'.png', bbox='tight',dpi=100)
```
